JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,

- Find the 10 countries with most projects
- Find the top 10 major project themes (using column 'mjtheme_namecode')
- In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# imports for Python, Pandas, Numpy

import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
#First step is to read json files as a panda data frame object

projects_json_df = pd.read_json('data/world_bank_projects.json')

#Inspection of column names and data read from the json file

projects_json_df.head()
```

```python
#Question 1: Find the 10 countries with most projects
#Since each row in the data corresponds to one project we can use the 'groupby' function to group the projects 
#by country name followed by the use of value_counts() to display the number of projects

countryprojects = projects_json_df.groupby('countryname').countryshortname.value_counts()

#We can now sort the results with the .sort() function to display the top 10 using the .head() function

topprojects = countryprojects.sort_values(ascending=False).head(15)
topprojects
```

```python
#Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')

#The first step is to normalize the data by loading json to pdata1

pdata1 = json.load((open('data/world_bank_projects.json')))

#The next step is to use the json_normalize() function to normalize the data read in to pdata1 
#while using the column 'mjtheme_namecode' and maintaining the values for the column, 'countryname'

pdatanorm = json_normalize(pdata1, 'mjtheme_namecode', ['countryname'])

#We can then group the data by 'code' and display the column 'name' while sorting the values in order
#to display the top 10 themes

pdatanorm.groupby('code').name.value_counts().sort_values(ascending=False).head(10)
```

```python
#Question 3: In Question 2 above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.

#Using the same steps as above we can normalize the data in to a data frame

pdatan = json_normalize(pdata1, 'mjtheme_namecode', ['countryname'])

#Using the .replace() function we can replace all empty values in the column 'name' with a Nan value 

pdatan['name'].replace('',np.nan, inplace=True)

#The next step is to gather all rows with same 'code' value in order to sort the data by the 'code' column

psorted = pdatan.sort_values(by='code')

#The next step is to use the fillna() function to replace the Nan values in the 'name' column and then use the
#'ffill' method of the 'fillna' function to populate the 'name' cells

psorted['name'].fillna(method='ffill', inplace=True)

#We can then repeat the group by steps to group by 'code' and display the revised top 10

psorted.groupby('code').name.value_counts().sort_values(ascending=False).head(10)
```
